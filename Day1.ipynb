# Task 5: Frequency Distribution

```
text = webtext.words(fileids = 'pirates.txt')
fd = nltk.FreqDist(text)
fd
```

#Task 6: Conditional Frequency Distribution

```
from nltk.probability import ConditionalFreqDist
cfd = ConditionalFreqDist((len(word), word) for word in text)
```

```
cfd[5]
```

#Task 7: Freq distribution and Conditional Freq distribution for the innaugural speech of a US President

```
Wash = '1789-Washington.txt'
Wash_sp = inaugural.words(fileids=Wash)
l = len(Wash_sp)
print('Length of George Washington\'s speech ',l)

print('Most frequently used words')
wfd = nltk.FreqDist(Wash_sp)
wfd.most_common(30)
```

```
print('Most sophisticated words')
wcfd = ConditionalFreqDist((len(word), word) for word in Wash_sp)
for i in range(10,20):
  print(wcfd[i])

wcfd[15]  
```

```
Trump = '2017-Trump.txt'
Trump_sp = inaugural.words(fileids=Trump)
l=len(Trump_sp)
print('Length of Trump\'s speech ',l)

print('Most frequently used words')
tfd = nltk.FreqDist(Trump_sp)
tfd.most_common(30)
```

```
print('Most sophisticated words')
tcfd = ConditionalFreqDist((len(word), word) for word in Trump_sp)
for i in range(10,20):
  print(tcfd[i])

tcfd[14]
```

The length of Trump's inaugural speech is longer than that of Washington. Also Washington used words with longer length with higher frequency, i.e., we can conclude that the vocabulary of Washington was better.
